```python
import numpy as np
import pandas as pd
import json
import pickle
import csv
from pandas.io.json import json_normalize
from py_translator import Translator
```

```python
file = './kshen/10-04-2017_fields.json'
```

```python
with open('./kshen/10-04-2017_fields.json') as f:
    d = json.load(f)
```

```python
df = json_normalize(d['features'])
```

```python
number = []
square = []
for i in np.asarray(df['properties.name']):
    temp = i.split(sep='-')
    
    number.append(float(temp[0]))
    square.append(float(temp[1]))
```

```python
df['properties.number'] = pd.Series(number)
df['properties.square'] = pd.Series(square)
df = df[['type','properties.number','properties.square','geometry.type','geometry.coordinates']]
df = df.sort_values(['properties.number'])
```

```python
with open('./kshen/summary_table.csv') as f:
    data = pd.read_csv(f, encoding='utf8')
```

```python
df['properties.crop']=pd.Series(np.nan)
df = df[['type','properties.number','geometry.type','properties.crop','geometry.coordinates']]
```

```python
my_translator = Translator()
```

```python
def translation(text):
    return my_translator.translate(text=text, dest='en').text
```

```python
df["properties.crop"]= (df["properties.number"]
                        .map(data.query('Отделение=="Кшень"')
                             .set_index("№ поля")["Культура в 2015 году"])
                       ).fillna('').apply(translation)
```

```python
df.columns = ['type','name','type','crop','coordinates']
df = df[['name','crop','coordinates']]
df.sort_values('name')
```

```python
def df_to_geojson(df, properties, coords):
    geojson = {'type':'FeatureCollection', 'features':[]}
    for _, row in df.iterrows():
        feature = {'type':'Feature',
                   'properties':{},
                   'geometry':{'type':'Polygon',
                               'coordinates':[]}}
        feature['geometry']['coordinates'] = (row[coords])
        for prop in properties:
            feature['properties'][prop] = row[prop]
        geojson['features'].append(feature)
    return geojson

def flatten(l): return flatten(l[0]) + (flatten(l[1:]) if len(l) > 1 else []) if type(l) is list else [l]
```

```python
properties = ['name','crop']
```

```python
geojson_dict = df_to_geojson(df, properties, coords = df.columns[-1])
geojson_str = json.dumps(geojson_dict, indent=2)
```

```python
output_filename = 'crop_data.json'
with open(output_filename, 'w') as output_file:
    output_file.write(geojson_str)
```
